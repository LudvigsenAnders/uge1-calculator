```python
import re

def split_on_non_numbers(s):
    """
    Splits a string into numbers (int or float) by removing non-numeric characters.
    Handles negative numbers and decimal points.
    """
    if not isinstance(s, str):
        raise TypeError("Input must be a string.")

    # Regex explanation:
    # -?         → optional minus sign
    # \d+        → one or more digits
    # (?:\.\d+)? → optional decimal part
    pattern = r"-?\d+(?:\.\d+)?"

    # Find all matches of numbers in the string
    matches = re.findall(pattern, s)

    # Convert to float if decimal present, else int
    result = [float(m) if '.' in m else int(m) for m in matches]
    return result


# Example usage:
examples = [
    "abc123def45.67gh-89xyz",
    "12 apples, 3.5 bananas, and -7.25 oranges",
    "no numbers here",
    "100-200+300.5"
]

for ex in examples:
    print(f"Input: {ex}")
    print("Output:", split_on_non_numbers(ex))
    print("-" * 40)

```
