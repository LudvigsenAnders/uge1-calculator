```python
import re
```

```python

def split_on_non_numbers(s):
    """
    Splits a string into numbers (int or float) by removing non-numeric characters.
    Handles negative numbers and decimal points.
    """
    if not isinstance(s, str):
        raise TypeError("Input must be a string.")

    # Regex explanation:
    # -?         → optional minus sign
    # \d+        → one or more digits
    # (?:\.\d+)? → optional decimal part
    #numbers_pattern = r"-?\d+(?:\.\d+)?"  #for extentions
    #operator_pattern = r"[+\-*/%^()]"      #for extentions


    numbers_pattern = r"\d+(?:\.\d+)?"
    operator_pattern = r"[+\-*/%^]"


    # Find all matches of numbers in the string
    num_matches = re.findall(numbers_pattern, s)
    # Find all matches of operators in the string
    operator_matches = re.findall(operator_pattern, s)

    # Convert to float
    num_result = [float(m) for m in num_matches]

    return num_result, operator_matches

```

```python
def add(num1, num2):
    return num1+num2

def subtract(num1,num2):
    return num1-num2

def multiply(num1,num2):
    return num1*num2

def divide(num1,num2):
    if num2 == 0.0:
        raise Exception("Sorry, your second input cannot be zero, divide by zero is not possible")
    else:
        return num1/num2
    
def power(num1,num2):
    return num1**num2

def n_root(num1,num2):
    return num1 ** divide(1,num2)
```

```python
def validate_input(num, opr):
    while True:
        try:
            if len(num) == 2:
                if opr:
                    break
                else:
                    raise ValueError("Invalid input! Please use a valid operator.")
            else:
                raise ValueError("Invalid input! Too many numbers. Please use this format '2+2'")
        except :
            raise ValueError("Invalid input! Please use this format '2+2' with a valid operator")
```

```python
def calculator():
    
    quit = False

    while not quit:
        text = input('enter your problem')

        
        if text == 'q':
            quit = True
        else:
            num, opr = split_on_non_numbers(text)
            validate_input(num, opr)
            
            result = None
            
            if opr[0] == '+':
                result = add(num[0],num[1])
            elif opr[0] == '-':
                result = subtract(num[0],num[1])
            elif  opr[0] == '/':
                result = divide(num[0],num[1])
            elif opr[0] == '*':
                result = multiply(num[0],num[1])
            elif opr[0] == '^':
                result = power(num[0],num[1])
            else:
                raise Exception("Sorry. How did we end here?")
            
            print(result)
    return
        
```

```python
calculator()
```
